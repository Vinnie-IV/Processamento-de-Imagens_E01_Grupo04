```python
image_path = '/content/media.jpg'

try:
    if not os.path.exists(image_path):
        raise FileNotFoundError(f"❌ Arquivo não encontrado: {image_path}")

    image = cv2.imread(image_path)
    if image is None:
        raise ValueError("❌ Erro ao ler a imagem. Verifique o formato ou o caminho do arquivo.")

    try:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    except cv2.error as e:
        raise ValueError("❌ Falha na conversão de cores. A imagem pode estar corrompida.") from e

    try:
        filtro = int(input("Digite o tamanho do filtro (número inteiro positivo): "))
        if filtro <= 0:
            raise ValueError("O tamanho do filtro deve ser um número positivo.")
    except ValueError:
        raise ValueError("❌ Entrada inválida. Digite um número inteiro positivo.")

    new_image = cv2.blur(image, (filtro, filtro))

    plt.figure(figsize=(11, 6))
    plt.subplot(121)
    plt.imshow(image)
    plt.title("Original")
    plt.xticks([]), plt.yticks([])

    plt.subplot(122)
    plt.imshow(new_image)
    plt.title(f"Aplicado a média ({filtro}x{filtro})")
    plt.xticks([]), plt.yticks([])

    plt.show()

except Exception as e:
    print(f"⚠️ Ocorreu um erro: {e}")
```

# MEDIANA



```python
image2_path = '/content/mediana.jpg'

try:
    if not os.path.exists(image2_path):
        raise FileNotFoundError(f"❌ Arquivo não encontrado: {image2_path}")

    image2 = cv2.imread(image2_path)
    if image2 is None:
        raise ValueError("❌ Erro ao ler a imagem. Verifique o formato ou o caminho do arquivo.")

    try:
        image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
    except cv2.error as e:
        raise ValueError("❌ Falha na conversão de cores. A imagem pode estar corrompida.") from e

    try:
        filtro2 = int(input("Digite o tamanho do filtro (ímpar e positivo): "))
        if filtro2 <= 0 or filtro2 % 2 == 0:
            raise ValueError("O filtro deve ser um número inteiro ímpar e positivo.")
    except ValueError:
        raise ValueError("❌ Entrada inválida. Digite um número inteiro ímpar e positivo.")

    try:
        median = cv2.medianBlur(image2, filtro2)
    except cv2.error as e:
        raise RuntimeError(f"❌ Erro ao aplicar o filtro de mediana: {e}")

    plt.figure(figsize=(11, 6))
    plt.subplot(121)
    plt.imshow(image2)
    plt.title("Original")
    plt.xticks([]), plt.yticks([])

    plt.subplot(122)
    plt.imshow(median)
    plt.title(f"Aplicado a mediana ({filtro2}x{filtro2})")
    plt.xticks([]), plt.yticks([])

    plt.show()

except Exception as e:
    print(f"⚠️ Ocorreu um erro: {e}")
```
